```python
import numpy as np
from PIL import Image, ImageOps
import PIL

image = Image.open("assets/images/tshirt.jpg")

```

```python
def resize_img(img):
    size = (28,28)
    gray_image = ImageOps.grayscale(img)
    img = gray_image.resize(size)
    image_array = np.array(img)
    return image_array


    
```

```python
resize_img(image)
```
